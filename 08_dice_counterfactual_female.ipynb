```python
# ============================================================
# 08_dice_counterfactual_female.ipynb
# DiCE Counterfactual Explanation — Female Cohort
#
# Paper: A Methodology for Interpretable Health Risk Management:
#        Integrating Counterfactual Explanations and
#        On-Premise Medical LLMs in the Insurance Industry
#
# Contents:
#   1. DiCE setup with XGBoost (final model)
#   2. Constraint definition (immutable / limited / free)
#   3. Counterfactual generation for non-diabetic high-risk group
#   4. Representative case analysis
#   5. Robustness analysis (XGBoost / RandomForest / LightGBM)
#   6. Export results for LLM pipeline (notebook 10)
#
# Note: Female robustness uses XGBoost / RandomForest / LightGBM
#       (top-3 accuracy models for female cohort, per notebook 04)
# ============================================================


# ─────────────────────────────────────────────
# Cell 1 | Library Imports
# ─────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

import dice_ml
from dice_ml import Dice

import joblib
import json
import os

os.makedirs('outputs/dice_female', exist_ok=True)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)

print("Libraries loaded successfully.")
try:
    print(f"DiCE version: {dice_ml.__version__}")
except AttributeError:
    import importlib.metadata
    try:
        print(f"DiCE version: {importlib.metadata.version('dice-ml')}")
    except Exception:
        print("DiCE version: unknown")


# ─────────────────────────────────────────────
# Cell 2 | Load Data & Pipeline
# ─────────────────────────────────────────────
df_female         = pd.read_csv('outputs/df_female.csv')
df_nd_highrisk    = pd.read_csv('outputs/df_female_nd_highrisk.csv')

pipeline_female   = joblib.load('outputs/models_female/final_pipeline_female.pkl')
model_xgb         = pipeline_female['model']
preprocessor      = pipeline_female['preprocessor']
thresh            = pipeline_female['threshold']
feature_col       = pipeline_female['feature_cols']
cat_col           = pipeline_female['cat_col']
num_col           = pipeline_female['num_col']
target_col        = pipeline_female['target_col']

with open('outputs/highrisk_cutoff_meta.json') as f:
    cutoff_meta = json.load(f)
prob_cutoff = cutoff_meta['female']['prob_cutoff']

print(f"Non-diabetic high-risk (female) : {len(df_nd_highrisk):,} cases")
print(f"High-risk prob cutoff           : {prob_cutoff:.4f}")
print(f"Model threshold                 : {thresh}")


# ─────────────────────────────────────────────
# Cell 3 | DiCE Data Object Setup
# ─────────────────────────────────────────────
df_train_dice = df_female[feature_col + [target_col]].copy()
df_train_dice[target_col] = df_train_dice[target_col].astype(int)

d = dice_ml.Data(
    dataframe           = df_train_dice,
    continuous_features = num_col,
    outcome_name        = target_col,
)
print("DiCE Data object created.")
print(f"  Continuous features : {num_col}")
print(f"  Categorical features: {cat_col}")


# ─────────────────────────────────────────────
# Cell 4 | DiCE Model Object (XGBoost)
# ─────────────────────────────────────────────
class XGBWrapper:
    """
    Wraps XGBoost + ColumnTransformer so DiCE can call
    predict_proba on raw (unprocessed) feature DataFrames.
    """
    def __init__(self, model, preprocessor, threshold):
        self.model        = model
        self.preprocessor = preprocessor
        self.threshold    = threshold

    def predict_proba(self, X):
        if isinstance(X, pd.DataFrame):
            X = self.preprocessor.transform(X)
        return self.model.predict_proba(X)

    def predict(self, X):
        probs = self.predict_proba(X)[:, 1]
        return (probs >= self.threshold).astype(int)


wrapped_model = XGBWrapper(model_xgb, preprocessor, thresh)

m = dice_ml.Model(
    model      = wrapped_model,
    backend    = 'sklearn',
    model_type = 'classifier',
)
print("DiCE Model object created (XGBoost + preprocessor wrapper).")


# ─────────────────────────────────────────────
# Cell 5 | Constraint Definition
# ─────────────────────────────────────────────
# ── Immutable features ──────────────────────
immutable_features = ['age']

# ── Permitted ranges (clinical plausibility) ──
# Female-specific adjustments:
#   · BMI    : safe lower bound 17.5 (slightly lower than male 18.5)
#   · HE_wc  : max 15 cm decrease
#   · HE_wt  : max 15 kg decrease
#   · N_SUGAR: 0 ~ 100g
#   · N_CHO  : 100 ~ 400g
#   · N_EN   : 1200 ~ 2800 kcal (lower upper bound for female)

def build_permitted_range(df, num_vars):
    ranges = {}
    ranges['HE_BMI']  = [max(17.5, round(df['HE_BMI'].min(), 1)),
                          round(df['HE_BMI'].max(), 1)]
    ranges['HE_wc']   = [max(55.0, round(df['HE_wc'].min(), 1)),
                          round(df['HE_wc'].max(), 1)]
    ranges['HE_wt']   = [max(35.0, round(df['HE_wt'].min(), 1)),
                          round(df['HE_wt'].max(), 1)]
    ranges['N_SUGAR'] = [0.0,    100.0]
    ranges['N_CHO']   = [100.0,  400.0]
    ranges['N_EN']    = [1200.0, 2800.0]
    ranges['age']     = [float(df['age'].min()),
                          float(df['age'].max())]
    return ranges

permitted_range    = build_permitted_range(df_female, num_col)
features_to_vary   = [f for f in feature_col if f not in immutable_features]

print("=== Permitted Ranges (Female — Clinical Plausibility) ===")
for k, v in permitted_range.items():
    print(f"  {k:12s} : {v}")
print(f"\nFeatures to vary : {len(features_to_vary)} / {len(feature_col)}")
print(f"Immutable        : {immutable_features}")


# ─────────────────────────────────────────────
# Cell 6 | DiCE Explainer Initialisation
# ─────────────────────────────────────────────
exp = Dice(d, m, method='random')
print("DiCE explainer initialised (method='random').")


# ─────────────────────────────────────────────
# Cell 7 | Counterfactual Generation Helper
# ─────────────────────────────────────────────
def generate_cf(query_instance, n_cf=4, desired_class=0,
                features_to_vary=features_to_vary,
                permitted_range=permitted_range,
                proximity_weight=0.5,
                diversity_weight=1.0,
                random_seed=42):
    cf_result = exp.generate_counterfactuals(
        query_instances  = query_instance,
        total_CFs        = n_cf,
        desired_class    = desired_class,
        features_to_vary = features_to_vary,
        permitted_range  = permitted_range,
        proximity_weight = proximity_weight,
        diversity_weight = diversity_weight,
        random_seed      = random_seed,
        verbose          = False,
    )
    cf_df     = cf_result.cf_examples_list[0].final_cfs_df
    orig_prob = float(
        wrapped_model.predict_proba(query_instance)[:, 1][0]
    )
    cf_probs  = [
        float(wrapped_model.predict_proba(
            cf_df.iloc[[i]][feature_col]
        )[:, 1][0])
        for i in range(len(cf_df))
    ]
    return cf_df, orig_prob, cf_probs


# ─────────────────────────────────────────────
# Cell 8 | Representative Case Analysis
# ─────────────────────────────────────────────
df_nd_sorted = df_nd_highrisk.sort_values(
    'pred_prob', ascending=False
).reset_index(drop=True)

rep_case     = df_nd_sorted.iloc[[0]][feature_col]
rep_prob_raw = float(df_nd_sorted.iloc[0]['pred_prob'])

print("=== Representative Case (Highest-Risk Non-Diabetic Female) ===")
print(f"  Predicted probability : {rep_prob_raw:.4f}")
print(rep_case[num_col + ['BS3_1', 'BD1_11', 'pa_aerobic',
                           'L_BR_FQ', 'edu', 'HE_obe']].T)


# ─────────────────────────────────────────────
# Cell 9 | Generate Counterfactuals
#          for Representative Case
# ─────────────────────────────────────────────
print("Generating counterfactuals for representative case...")

cf_df_rep, orig_prob_rep, cf_probs_rep = generate_cf(
    query_instance = rep_case,
    n_cf           = 4,
)

print(f"\n  Original risk probability : {orig_prob_rep:.4f}")
print(f"  Counterfactual probs      : "
      f"{[round(p, 4) for p in cf_probs_rep]}")

# Build comparison table
comp_rows = []
orig_row  = rep_case.iloc[0].to_dict()
orig_row['Type']      = 'Original'
orig_row['Pred_Prob'] = round(orig_prob_rep, 4)
comp_rows.append(orig_row)

for i, (_, cf_row) in enumerate(cf_df_rep.iterrows()):
    row               = cf_row[feature_col].to_dict()
    row['Type']       = f'CF_{i+1}'
    row['Pred_Prob']  = round(cf_probs_rep[i], 4)
    comp_rows.append(row)

df_comp    = pd.DataFrame(comp_rows).set_index('Type')
orig_vals  = df_comp.loc['Original', feature_col]
changed    = [
    c for c in feature_col
    if not all(df_comp.loc[df_comp.index != 'Original', c] == orig_vals[c])
]
show_cols  = changed + ['Pred_Prob']

print("\n=== Table: Representative Case — Counterfactual Comparison ===")
print(df_comp[show_cols].to_string())

df_comp.to_csv('outputs/dice_female/rep_case_cf.csv')
print("\nSaved: outputs/dice_female/rep_case_cf.csv")


# ─────────────────────────────────────────────
# Cell 10 | Figure 1 — Counterfactual
#           Feature Change Visualization
# ─────────────────────────────────────────────
changed_num = [c for c in changed if c in num_col]

if changed_num:
    fig, axes = plt.subplots(
        1, len(changed_num),
        figsize=(4 * len(changed_num), 4)
    )
    if len(changed_num) == 1:
        axes = [axes]

    colors = ['#DD8452', '#4C72B0', '#55A868', '#C44E52']

    for ax, col in zip(axes, changed_num):
        orig_val = float(df_comp.loc['Original', col])
        cf_vals  = [float(df_comp.loc[f'CF_{i+1}', col]) for i in range(4)]
        labels   = ['Original'] + [f'CF_{i+1}' for i in range(4)]
        vals     = [orig_val] + cf_vals
        bar_cols = ['#aaaaaa'] + colors

        bars = ax.bar(labels, vals, color=bar_cols, alpha=0.85, width=0.5)
        for bar, val in zip(bars, vals):
            ax.text(bar.get_x() + bar.get_width()/2,
                    bar.get_height() + 0.3,
                    f'{val:.1f}', ha='center', va='bottom', fontsize=9)
        ax.set_title(col, fontsize=11)
        ax.set_ylabel('Value')
        ax.set_xticklabels(labels, rotation=20, ha='right', fontsize=9)

    fig.suptitle(
        'Figure 1: Feature Changes — Representative Case (Female)\n'
        f'Original Risk: {orig_prob_rep:.3f}  →  '
        f'CF Risk: {[round(p, 3) for p in cf_probs_rep]}',
        fontsize=11
    )
    plt.tight_layout()
    plt.savefig('outputs/dice_female/fig_cf_rep_case.png',
                bbox_inches='tight')
    plt.show()
    print("Saved: outputs/dice_female/fig_cf_rep_case.png")
else:
    print("No continuous feature changes detected.")


# ─────────────────────────────────────────────
# Cell 11 | Batch Counterfactual Generation
# ─────────────────────────────────────────────
BATCH_N = min(50, len(df_nd_sorted))
print(f"Generating CFs for top {BATCH_N} non-diabetic high-risk females...")

batch_results = []

for idx in range(BATCH_N):
    query = df_nd_sorted.iloc[[idx]][feature_col]
    try:
        cf_df, orig_p, cf_ps = generate_cf(
            query_instance = query,
            n_cf           = 4,
            random_seed    = 42 + idx,
        )
        for i, (_, cf_row) in enumerate(cf_df.iterrows()):
            row = {
                'case_id'       : idx,
                'case_type'     : 'CF',
                'cf_rank'       : i + 1,
                'orig_prob'     : round(orig_p, 4),
                'cf_prob'       : round(cf_ps[i], 4),
                'prob_reduction': round(orig_p - cf_ps[i], 4),
            }
            orig_vals_b = query.iloc[0].to_dict()
            for col in feature_col:
                row[f'orig_{col}'] = orig_vals_b[col]
                row[f'cf_{col}']   = cf_row[col]
                row[f'delta_{col}']= (
                    cf_row[col] - orig_vals_b[col]
                    if col in num_col else None
                )
            batch_results.append(row)

        if (idx + 1) % 10 == 0:
            print(f"  Processed {idx+1}/{BATCH_N}...")
    except Exception as e:
        print(f"  [Warning] Case {idx} failed: {e}")
        continue

df_batch = pd.DataFrame(batch_results)
df_batch.to_csv('outputs/dice_female/batch_cf_female.csv', index=False)
print(f"\nBatch CFs saved: {len(df_batch)} rows")
print("Saved: outputs/dice_female/batch_cf_female.csv")


# ─────────────────────────────────────────────
# Cell 12 | Figure 2 — Probability Reduction
#           Distribution (Batch Results)
# ─────────────────────────────────────────────
df_best_cf = (
    df_batch.groupby('case_id')
    .apply(lambda x: x.loc[x['cf_prob'].idxmin()])
    .reset_index(drop=True)
)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df_best_cf['orig_prob'], bins=20,
             color='#DD8452', alpha=0.8, edgecolor='white')
axes[0].set_title('Original Risk Probability Distribution\n'
                  '(Non-diabetic High-Risk Females)', fontsize=11)
axes[0].set_xlabel('Predicted Probability')
axes[0].set_ylabel('Count')
axes[0].axvline(df_best_cf['orig_prob'].mean(), color='red',
                linestyle='--',
                label=f"Mean={df_best_cf['orig_prob'].mean():.3f}")
axes[0].legend()

axes[1].hist(df_best_cf['prob_reduction'], bins=20,
             color='#DD8452', alpha=0.8, edgecolor='white')
axes[1].set_title('Risk Reduction via Best Counterfactual\n'
                  f'(Female, n={len(df_best_cf)})', fontsize=11)
axes[1].set_xlabel('Probability Reduction (Original − CF)')
axes[1].set_ylabel('Count')
axes[1].axvline(df_best_cf['prob_reduction'].mean(), color='red',
                linestyle='--',
                label=f"Mean={df_best_cf['prob_reduction'].mean():.3f}")
axes[1].legend()

plt.tight_layout()
plt.savefig('outputs/dice_female/fig_prob_reduction.png', bbox_inches='tight')
plt.show()
print("Saved: outputs/dice_female/fig_prob_reduction.png")

print(f"\n  Avg original prob   : {df_best_cf['orig_prob'].mean():.4f}")
print(f"  Avg CF prob         : {df_best_cf['cf_prob'].mean():.4f}")
print(f"  Avg reduction       : {df_best_cf['prob_reduction'].mean():.4f}")


# ─────────────────────────────────────────────
# Cell 13 | Figure 3 — Most Frequently Changed
#           Features (Batch Analysis)
# ─────────────────────────────────────────────
change_counts = {}
for col in num_col:
    delta_col = f'delta_{col}'
    if delta_col in df_batch.columns:
        change_counts[col] = (df_batch[delta_col].abs() > 0.01).sum()

for col in cat_col:
    orig_c = f'orig_{col}'
    cf_c   = f'cf_{col}'
    if orig_c in df_batch.columns and cf_c in df_batch.columns:
        change_counts[col] = (df_batch[orig_c] != df_batch[cf_c]).sum()

change_series = pd.Series(change_counts).sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(8, 7))
bars = ax.barh(change_series.index, change_series.values,
               color='#DD8452', alpha=0.85)
ax.set_xlabel('Number of CFs where feature was changed', fontsize=11)
ax.set_title('Figure 3: Most Frequently Modified Features\n'
             f'(Female, Batch N={BATCH_N}, 4 CFs each)', fontsize=12)
for bar, val in zip(bars, change_series.values):
    ax.text(bar.get_width() + 0.5,
            bar.get_y() + bar.get_height()/2,
            str(val), va='center', fontsize=9)
plt.tight_layout()
plt.savefig('outputs/dice_female/fig_feature_change_freq.png',
            bbox_inches='tight')
plt.show()
print("Saved: outputs/dice_female/fig_feature_change_freq.png")


# ─────────────────────────────────────────────
# Cell 14 | Robustness Analysis
#           XGBoost / RandomForest / LightGBM
#           (Top-3 accuracy models — Female cohort)
# ─────────────────────────────────────────────
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
try:
    from pingouin import intraclass_corr
    PINGOUIN_AVAILABLE = True
except ImportError:
    PINGOUIN_AVAILABLE = False
    print("[Warning] pingouin not installed. ICC will be skipped.")
    print("  Install: pip install pingouin")

print("=== Robustness Analysis: Multi-Model Consistency ===")
model_rf  = joblib.load('outputs/models_female/rf_female.pkl')
model_lgb = joblib.load('outputs/models_female/lgb_female.pkl')

robust_models = {
    'XGBoost'     : (model_xgb, preprocessor),
    'RandomForest': (model_rf,  preprocessor),
    'LightGBM'    : (model_lgb, preprocessor),
}

# Build DiCE explainers for each model
robust_explainers = {}
for name, (mdl, prep) in robust_models.items():
    w   = XGBWrapper(mdl, prep, thresh)
    m_  = dice_ml.Model(
        model=w, backend='sklearn', model_type='classifier'
    )
    robust_explainers[name] = Dice(d, m_, method='random')

print("Robustness explainers ready.")

# 10 representative cases
robust_cases = df_nd_sorted.iloc[:10][feature_col].reset_index(drop=True)
robust_results = {name: [] for name in robust_models}

for idx in range(len(robust_cases)):
    query = robust_cases.iloc[[idx]]
    for name, exp_ in robust_explainers.items():
        try:
            cf_r = exp_.generate_counterfactuals(
                query_instances  = query,
                total_CFs        = 4,
                desired_class    = 0,
                features_to_vary = features_to_vary,
                permitted_range  = permitted_range,
                random_seed      = 42,
                verbose          = False,
            )
            cf_df_ = cf_r.cf_examples_list[0].final_cfs_df
            mdl_, prep_ = robust_models[name]
            probs_ = [
                float(mdl_.predict_proba(
                    prep_.transform(cf_df_.iloc[[i]][feature_col])
                )[:, 1][0])
                for i in range(len(cf_df_))
            ]
            robust_results[name].append({
                'case_id' : idx,
                'cf_probs': probs_,
                'best_cf' : min(probs_),
                'cf_df'   : cf_df_,
            })
        except Exception as e:
            print(f"  [Warning] {name} case {idx}: {e}")
            robust_results[name].append({
                'case_id': idx, 'cf_probs': [None]*4,
                'best_cf': None, 'cf_df': None,
            })


# ─────────────────────────────────────────────
# Cell 15 | Robustness — ICC
# ─────────────────────────────────────────────
icc_rows = []
for name, results in robust_results.items():
    for r in results:
        if r['best_cf'] is not None:
            icc_rows.append({
                'case_id': r['case_id'],
                'rater'  : name,
                'score'  : r['best_cf'],
            })

df_icc = pd.DataFrame(icc_rows)

if PINGOUIN_AVAILABLE and len(df_icc) > 0:
    try:
        icc_result = intraclass_corr(
            data    = df_icc,
            targets = 'case_id',
            raters  = 'rater',
            ratings = 'score',
        )
        icc_val = icc_result.set_index('Type').loc['ICC2', 'ICC']
        print(f"\nICC2 (two-way mixed, absolute agreement) = {icc_val:.3f}")
        print(icc_result[['Type','ICC','CI95%','pval']].to_string(index=False))
    except Exception as e:
        print(f"  [Warning] ICC computation failed: {e}")
else:
    # Manual consistency fallback
    pivot = df_icc.pivot(index='case_id', columns='rater', values='score')
    print("\nManual consistency (correlation matrix):")
    print(pivot.corr().round(3))


# ─────────────────────────────────────────────
# Cell 16 | Robustness — Feature Consistency
# ─────────────────────────────────────────────
print("\n=== Robustness: Recommended Feature Consistency ===")
consistency_rows = []

for idx in range(len(robust_cases)):
    top_features = {}
    for name in robust_models:
        r = robust_results[name][idx]
        if r['cf_df'] is None:
            continue
        cf_df_   = r['cf_df']
        orig_row = robust_cases.iloc[idx]
        changes  = {}
        for col in feature_col:
            cf_vals_ = cf_df_[col].values
            orig_val = orig_row[col]
            if col in num_col:
                avg_delta = np.mean(np.abs(cf_vals_ - orig_val))
            else:
                avg_delta = float(np.mean(cf_vals_ != orig_val))
            if avg_delta > 0.01:
                changes[col] = avg_delta
        if changes:
            top_features[name] = set(
                sorted(changes, key=changes.get, reverse=True)[:3]
            )

    if len(top_features) >= 2:
        names    = list(top_features.keys())
        overlaps = []
        for i in range(len(names)):
            for j in range(i+1, len(names)):
                a, b    = top_features[names[i]], top_features[names[j]]
                overlap = len(a & b) / len(a | b) if (a | b) else 0
                overlaps.append(overlap)
        consistency_rows.append({
            'case_id'          : idx,
            'consistency_score': round(np.mean(overlaps), 3),
            'models_compared'  : str(names),
        })

df_consistency = pd.DataFrame(consistency_rows)
mean_cons = df_consistency['consistency_score'].mean()
std_cons  = df_consistency['consistency_score'].std()

print(f"\n  Overall consistency : {mean_cons:.3f} ± {std_cons:.3f}")
print(f"  Range               : "
      f"{df_consistency['consistency_score'].min():.3f} – "
      f"{df_consistency['consistency_score'].max():.3f}")
print(df_consistency.to_string(index=False))

df_consistency.to_csv(
    'outputs/dice_female/robustness_consistency.csv', index=False
)
print("\nSaved: outputs/dice_female/robustness_consistency.csv")


# ─────────────────────────────────────────────
# Cell 17 | Figure 4 — Robustness: Best CF Prob
#           Comparison Across Models
# ─────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))

colors_rob = {
    'XGBoost'     : '#e41a1c',
    'RandomForest': '#377eb8',
    'LightGBM'    : '#4daf4a',
}
x       = np.arange(len(robust_cases))
width   = 0.25
offsets = [-width, 0, width]

for (name, results), offset in zip(robust_results.items(), offsets):
    best_probs = [
        r['best_cf'] if r['best_cf'] is not None else np.nan
        for r in results
    ]
    ax.bar(x + offset, best_probs, width,
           label=name, color=colors_rob[name], alpha=0.8)

ax.axhline(thresh, color='black', linestyle='--', lw=1.5,
           label=f'Classification threshold ({thresh})')
ax.set_xticks(x)
ax.set_xticklabels(
    [f'Case {i}' for i in range(len(robust_cases))],
    rotation=30, ha='right', fontsize=9
)
ax.set_ylabel('Best CF Predicted Probability')
ax.set_title('Figure 4: Robustness Analysis — Best CF Risk by Model\n'
             '(Female, 10 Representative Cases)', fontsize=11)
ax.legend(fontsize=9)
ax.set_ylim([0, 1])
plt.tight_layout()
plt.savefig('outputs/dice_female/fig_robustness.png', bbox_inches='tight')
plt.show()
print("Saved: outputs/dice_female/fig_robustness.png")


# ─────────────────────────────────────────────
# Cell 18 | Export LLM Input Payload
#           (Top-N cases → notebook 10)
# ─────────────────────────────────────────────
LLM_EXPORT_N = min(20, BATCH_N)

llm_payloads = []
for idx in range(LLM_EXPORT_N):
    case_batch = df_batch[df_batch['case_id'] == idx]
    if case_batch.empty:
        continue

    orig_row = df_nd_sorted.iloc[idx]
    best_cf  = case_batch.loc[case_batch['cf_prob'].idxmin()]

    changed_feats = []
    for col in num_col:
        delta = best_cf[f'delta_{col}']
        if abs(delta) > 0.01:
            changed_feats.append({
                'feature'       : col,
                'original_value': round(float(best_cf[f'orig_{col}']), 2),
                'cf_value'      : round(float(best_cf[f'cf_{col}']), 2),
                'delta'         : round(float(delta), 2),
            })
    for col in cat_col:
        orig_v = best_cf[f'orig_{col}']
        cf_v   = best_cf[f'cf_{col}']
        if orig_v != cf_v:
            changed_feats.append({
                'feature'       : col,
                'original_value': orig_v,
                'cf_value'      : cf_v,
                'delta'         : None,
            })

    payload = {
        'case_id'         : int(idx),
        'gender'          : 'female',
        'age'             : int(orig_row['age']),
        'original_prob'   : round(float(best_cf['orig_prob']), 4),
        'cf_prob'         : round(float(best_cf['cf_prob']), 4),
        'prob_reduction'  : round(float(best_cf['prob_reduction']), 4),
        'changed_features': changed_feats,
        'static_features' : {
            col: orig_row[col]
            for col in ['HE_BMI', 'HE_wc', 'age',
                        'BS3_1', 'BD1_11', 'pa_aerobic']
            if col in orig_row.index
        },
    }
    llm_payloads.append(payload)

with open('outputs/dice_female/llm_input_female.json', 'w') as f:
    json.dump(llm_payloads, f, indent=2, default=str)

print(f"LLM input payloads exported: {len(llm_payloads)} cases")
print("Saved: outputs/dice_female/llm_input_female.json")

print("\n=== Preview: Case 0 LLM Payload ===")
print(json.dumps(llm_payloads[0], indent=2, default=str))


# ─────────────────────────────────────────────
# Cell 19 | Summary Report
# ─────────────────────────────────────────────
print("=" * 65)
print("DiCE COUNTERFACTUAL SUMMARY — FEMALE COHORT")
print("=" * 65)
print(f"\n[Input]")
print(f"  Non-diabetic high-risk cases : {len(df_nd_highrisk):,}")
print(f"  Batch processed              : {BATCH_N}")
print(f"  CFs per case                 : 4")

print(f"\n[Counterfactual Quality]")
print(f"  Avg original prob   : {df_best_cf['orig_prob'].mean():.4f}")
print(f"  Avg best CF prob    : {df_best_cf['cf_prob'].mean():.4f}")
print(f"  Avg risk reduction  : {df_best_cf['prob_reduction'].mean():.4f}")
print(f"  Success rate        : "
      f"{(df_best_cf['cf_prob'] < thresh).mean()*100:.1f}%"
      f" (CF below threshold)")

print(f"\n[Robustness]")
print(f"  Consistency score   : {mean_cons:.3f} ± {std_cons:.3f}")
print(f"  Models compared     : XGBoost, RandomForest, LightGBM")

print(f"\n[Key Difference from Male Cohort]")
print(f"  · Robustness models: XGBoost / RandomForest / LightGBM")
print(f"    (Male used CatBoost; Female top-3 differs per notebook 04)")
print(f"  · Energy intake upper bound: 2800 kcal (vs 3500 for male)")
print(f"  · BMI lower bound: 17.5 (vs 18.5 for male)")

print(f"\n[Methodological Note]")
print(f"  All CFs are statistical association-based hypothetical")
print(f"  scenarios. They do not imply causal relationships.")
print(f"  Cross-sectional data limits causal inference.")
print(f"  Reverse causality possible for lifestyle variables.")

print(f"\n[Outputs]")
for f_ in ['rep_case_cf.csv', 'batch_cf_female.csv',
           'robustness_consistency.csv', 'llm_input_female.json',
           'fig_cf_rep_case.png', 'fig_prob_reduction.png',
           'fig_feature_change_freq.png', 'fig_robustness.png']:
    print(f"  outputs/dice_female/{f_}")

print(f"\n{'='*65}")
print("Proceed to 09_robustness_analysis.ipynb")
```
